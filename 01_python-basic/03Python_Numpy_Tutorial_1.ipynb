###Random   
*   np.random.reandint 균일 분포의 정수 난수 1개 생성
*   np.random.rand 0부터 1사이의 군일 분포에서 난수 matrix array생성
*   np.random.randn 가우시안 표준 정규 분포에서 난수 matrix array생성






```
np.random.seed(0)
p = np.array([0.1, 0.0, 0.7, 0.2])
```

```
for i in range(10):
 index = np.random.choice([0, 1, 2, 3], p = p.ravel())
 print(i, index)
```
