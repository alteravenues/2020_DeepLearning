#간단한 ForWard 연산 해보기

*   bias는 일단 무시하고 진행




입력 층 (input layer)

```
X = np.array([1.0, 0.5])
```

```
X.shape
```

```
X
```

입력의 2개가 다음 layer에 있는 3개와 연결

```
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
```

```
W1.shape
```

```
W1
```

곱해보기 (dot-product)

```
a1 = np.dot(X, W1)
```

```
a1.shape
```

```
a1
```

```
1.0 * 0.1 + 0.5 * 0.2
```

#간단한 Forward 연산 해보기

*   bias 고려




```
b1 = np.array([0.1, 0.2, 0.3])
```

```
a1 = np.dot(X, W1) + b1
```

```
print(a1)
```
