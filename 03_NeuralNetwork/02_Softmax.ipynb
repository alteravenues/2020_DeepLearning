```
import numpy as np
```

```
a = np.array([0.3, 0.9, 5.0])
```

```
a = np.array([0.3, -0.9, 5.0])
```

```
a
```

```
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```
y = softmax(a)
```

```
print(y)
```

```
np.sum(y)
```

```
a = np.array([1010, 1000, 990]) #큰 수 준비
```

```
y = softmax(a)
```

```
print(y)  #제대로 계산되지 않음.
```

```
def stable_softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)   #개선
    sum_exp_a = sum(np.exp(a-c))
    y = exp_a/sum_exp_a
    return y
```

```
y = stable_softmax(a)
```

```
print(y)
[9.99954600e-01 4.53978686e-05 2.06106005e-09]
```

```
np.sum(y)   # 마찬가지로 더하면 1을 출력!
```
